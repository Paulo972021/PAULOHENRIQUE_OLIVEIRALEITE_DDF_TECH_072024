Expectation da Base olist_customers_dataset

```python
import great_expectations as gx
import pandas as pd

#Lendo o arquivo
df = pd.read_csv( 'C:/Users/paulo.henrique/Documents/olist/olist_customers_dataset.csv', )

#Lendo o dataframe
df.head()
df_gx = gx.from_pandas(df)

print(type(df_gx))
df_gx.head()


# Expectativa 1
df_gx.expect_column_values_to_be_of_type(column='customer_id', type_='str')

df_gx.expect_column_values_to_be_unique(column='customer_id')

df_gx.expect_column_values_to_not_be_null(column='customer_id')

# Expectativa 2
df_gx.expect_column_values_to_be_of_type(column='customer_unique_id', type_='str')

df_gx.expect_column_values_to_be_unique(column='customer_unique_id')

df_gx.expect_column_values_to_not_be_null(column='customer_unique_id')

# Expectativa 3
df_gx.expect_column_values_to_be_of_type(column='customer_zip_code_prefix', type_='int')

df_gx.expect_column_values_to_not_be_null(column='customer_zip_code_prefix')

# Expectativa 4
df_gx.expect_column_values_to_be_of_type(column='customer_city', type_='str')
df_gx.expect_column_values_to_not_be_null(column='customer_city')

results = df_gx.validate()
print(results)


validation_results = []


# Iterar sobre os resultados e extrair informaes
for result in results["results"]:
    expectation_type = result["expectation_config"]["expectation_type"]
    column = result["expectation_config"]["kwargs"].get("column")
    success = result["success"]
    validation_results.append({
        "Expectation": expectation_type,
        "Column": column,
        "Success": success
    })

# Criar um DataFrame com os resultados
df_results = pd.DataFrame(validation_results)

df_results.to_csv('C:/Users/paulo.henrique/Documents/olist/list_customers_DQ.csv', index=False, encoding='utf-8')
```
